```sql
--This question told total profit each by by data story by sales_amount is total_price & price is price that products per piece
SELECT *,
  CONCAT(ROUND((total_profit/total_cost)*100,2), ' %') AS percentage
FROM(
  SELECT date_trunc('month', order_date)::DATE AS order_dates,
    SUM(sales_amount) AS total_prices,
    SUM(quantity * cost) AS total_cost,
    (SUM(sales_amount) - SUM(quantity * cost)) AS total_profit
  FROM workspace.salesdb.fact_sales
  JOIN workspace.salesdb.dim_products ON  
    dim_products.product_key = fact_sales.product_key
  GROUP BY 1
) t
ORDER BY order_dates
```
