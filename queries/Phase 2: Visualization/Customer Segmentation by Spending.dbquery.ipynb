```sql
--this query for seperate customer group by spending by we assume 
--if our customer spent more than 4000$ -> VIP PLATINUM
--if our customer spent between 3000$ - 3999$ -> VIP GOLD
--if our customer spent between 2000$ - 2999$ -> VIP SILVER
--if our customer spent between 1000$ - 1999$ -> VIP BRONZE
--if our customer spent less than 1000$ -> STANDARD
--and show how many customer in each group
SELECT customer_tier,
  COUNT(customer_number) AS number_of_customer
FROM (
  SELECT customer_number,
    SUM(sales_amount) AS Total_Spent,
    CASE
      WHEN SUM(sales_amount) >= 4000 THEN 'VIP PLATINUM'
      WHEN SUM(sales_amount) BETWEEN 3000 AND 3999 THEN 'VIP GOLD'
      WHEN SUM(sales_amount) BETWEEN 2000 AND 2999 THEN 'VIP SILVER'
      ELSE 'VIP BRONZE'
    END AS customer_tier
  FROM workspace.salesdb.dim_customers
  JOIN workspace.salesdb.fact_sales ON
    workspace.salesdb.dim_customers.customer_key = workspace.salesdb.fact_sales.customer_key
  GROUP BY 1
)
GROUP BY 1
ORDER BY 
  CASE
    WHEN customer_tier = 'VIP PLATINUM' THEN 1
    WHEN customer_tier = 'VIP GOLD' THEN 2
    WHEN customer_tier = 'VIP SILVER' THEN 3
    ELSE 4
  END DESC
```
