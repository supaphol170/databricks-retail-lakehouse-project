```sql
--This query wanna show Top 10 customer by sales, that's mean most top 10 users as bought
SELECT customer_number,
  SUM(sales_amount) AS total_sales
FROM workspace.salesdb.dim_customers
JOIN workspace.salesdb.fact_sales ON
  workspace.salesdb.dim_customers.customer_key = workspace.salesdb.fact_sales.customer_key
GROUP BY 1
ORDER BY 2 DESC
LIMIT 10;
```
